```python
# Import libraries
import pandas as pd
import openpyxl
from ydata_profiling import ProfileReport
```

### Importing Datasets

```python
df_item = pd.read_csv("C:/Users/Jeffs/Documents/PADS Insper/3T/order_items.csv")
df_reviews = pd.read_csv("C:/Users/Jeffs/Documents/PADS Insper/3T/order_reviews.csv")
df_orders = pd.read_csv("C:/Users/Jeffs/Documents/PADS Insper/3T/orders.csv")
df_products = pd.read_csv("C:/Users/Jeffs/Documents/PADS Insper/3T/products.csv")
df_geolocation = pd.read_csv("C:/Users/Jeffs/Documents/PADS Insper/3T/geolocation.csv")
df_sellers = pd.read_csv("C:/Users/Jeffs/Documents/PADS Insper/3T/sellers.csv")
df_order_pay = pd.read_csv("C:/Users/Jeffs/Documents/PADS Insper/3T/order_payments.csv")
df_customers = pd.read_csv("C:/Users/Jeffs/Documents/PADS Insper/3T/customers.csv")
df_category = pd.read_csv("C:/Users/Jeffs/Documents/PADS Insper/3T/product_category_name_translation.csv")
```

### Merging Datasets

```python
df_olist = df_orders.merge(df_item, on='order_id', how='left')
df_olist = df_olist.merge(df_order_pay, on='order_id', how='outer', validate='m:m')
df_olist = df_olist.merge(df_reviews, on='order_id', how='outer')
df_olist = df_olist.merge(df_products, on='product_id', how='outer')
df_olist = df_olist.merge(df_customers, on='customer_id', how='outer')
df_olist = df_olist.merge(df_sellers, on='seller_id', how='outer')
# sem geolocation
#  df_olist = df_olist.merge(df_customers, on='zip_code_prefix', how='outer')

print(df_olist.shape)
```
